```python
# Date - 1st June 2022

import netCDF4 as nc
import matplotlib.pyplot as plt

# Path to your netCDF file
netcdf_file_path = "D:/GPM/Final/Original/Daily/20220601.nc"

# Open the netCDF file
dataset = nc.Dataset(netcdf_file_path, 'r')

# Read the variables
lon = dataset.variables['lon'][:]
lat = dataset.variables['lat'][:]
precipitation = dataset.variables['precipitationCal'][0, :, :]  # Assuming first time index

# Close the dataset
dataset.close()

# Plotting
plt.figure(figsize=(10, 6))
plt.pcolormesh(lon, lat, precipitation, shading='auto')
plt.colorbar(label='Precipitation (mm)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Daily Precipitation')
plt.show()
```

```python
#IMD netcdf.

import netCDF4 as nc

# Replace with your netCDF file path
file_path = "D:/GPM/IMD 2022/RF25_ind2022_rfp25.nc"

# Open the netCDF file
dataset = nc.Dataset(file_path, 'r')

print("Dimensions:")
for dimname in dataset.dimensions:
    print("   ", dimname, dataset.dimensions[dimname].size)

print("\nVariables:")
for varname in dataset.variables:
    print("   ", varname, dataset.variables[varname].ncattrs())

print("\nAttributes:")
for attr in dataset.ncattrs():
    print("   ", attr, getattr(dataset, attr))

# Close the dataset
dataset.close()
```
