```python
"""Find missing files from folder"""
import datetime
import os
import re

input_folder = r"D:\GPM\Final"  # Replace with your path

# Define the start and end dates for your dataset
start_date = datetime.datetime(2022, 6, 1, 0, 0)  # Example start date
end_date = datetime.datetime(2022, 6, 7, 23, 30)  # Example end date

current_date = start_date
missing_files = []

while current_date <= end_date:
    # Construct the expected file name pattern
    date_str = current_date.strftime("%Y%m%d")
    time_str = current_date.strftime("%H%M%S")
    file_pattern = f"3B-HHR.MS.MRG.3IMERG.{date_str}-S{time_str}-E.*\.nc4"

    # Check if a file matching this pattern exists in the folder
    file_found = False
    for file_name in os.listdir(input_folder):
        if re.match(file_pattern, file_name):
            file_found = True
            break

    if not file_found:
        missing_files.append(f"Expected pattern: {file_pattern}")

    # Increment the current date by 30 minutes
    current_date += datetime.timedelta(minutes=30)

# Report the results
if missing_files:
    print(f"Missing files or patterns ({len(missing_files)}):")
    for file in missing_files:
        print(file)
else:
    print("No missing files detected.")
```

```python
"""Re-names files to date time specifications"""
import os
import re

input_folder = r"D:\GPM\Final\New Folder"  # Replace with your path

# List all files in the directory
file_list = os.listdir(input_folder)

for file_name in file_list:
    # Match the filename with the known pattern
    match = re.match(r"3B-HHR\.MS\.MRG\.3IMERG\.(\d{8})-S(\d{6})-E\d{6}\.\d{4}\.V\d{2}A\.HDF5\.nc4", file_name)
    if match:
        # Extract the date and start time
        date_str, time_str = match.groups()
        new_file_name = f"{date_str}-S{time_str[:4]}.nc"  # New filename format

        # Full path for the old and new files
        old_file_path = os.path.join(input_folder, file_name)
        new_file_path = os.path.join(input_folder, new_file_name)

        # Rename the file
        os.rename(old_file_path, new_file_path)
        #print(f"Renamed '{file_name}' to '{new_file_name}'")

print("File renaming completed.")
```

```python
import numpy as np
import datetime
import netCDF4 as nc
import os
import re

def process_imerg_data(start_date, end_date, input_folder, output_folder):
    # Convert start_date and end_date to datetime objects
    start_date = datetime.datetime.strptime(start_date, "%Y%m%d")
    end_date = datetime.datetime.strptime(end_date, "%Y%m%d")

    # Get the list of files in the input folder
    LF = os.listdir(input_folder)

    # Iterate over each day in the date range
    current_date = start_date
    while current_date <= end_date:
        start_time = current_date.replace(hour=3, minute=0)  # starts at 3.00 am GMT
        end_time = start_time + datetime.timedelta(days=1) - datetime.timedelta(minutes=30)
        
        # Find the files within the specified time range
        LF2 = []
        for file_name in LF:
            match = re.match(r"\d{8}-S\d{4}\.nc", file_name)
            if match:
                file_date = datetime.datetime.strptime(file_name[:8], "%Y%m%d")
                if start_time.date() == file_date.date():
                    LF2.append(file_name)

        # Process the selected files
        precip3D = []
        for file_name in LF2:
            file_path = os.path.join(input_folder, file_name)
            try:
                dataset = nc.Dataset(file_path)
                for var_name, variable in dataset.variables.items():
                    if variable.dimensions == ('time', 'lon', 'lat'):
                        data = variable[:]
                        precip3D.append(data)
                dataset.close()
            except OSError as e:
                print("Error opening file:", file_path)
                print(e)
                continue

        # Aggregate data
        if precip3D:
            precip3D = np.concatenate(precip3D, axis=0)
            precip3D_sum = np.sum(precip3D, axis=0)

            # Create and write to a new NetCDF file
            date_str = current_date.strftime("%Y%m%d")
            output_file_name = f"{date_str}.nc"
            output_file = os.path.join(output_folder, output_file_name)
            with nc.Dataset(output_file, "w", format="NETCDF4") as dataset:
                lon_dim, lat_dim = precip3D_sum.shape[1], precip3D_sum.shape[2]
                dataset.createDimension("time", 1)
                dataset.createDimension("lon", lon_dim)
                dataset.createDimension("lat", lat_dim)

                time_var = dataset.createVariable("time", np.float64, ("time",))
                lon_var = dataset.createVariable("lon", np.float32, ("lon",))
                lat_var = dataset.createVariable("lat", np.float32, ("lat",))
                precip_var = dataset.createVariable("precipitation", np.float32, 
                                                    ("time", "lon", "lat"), 
                                                    fill_value=-9999.900390625)

                # Dummy values for lon and lat, replace with actual values
                lon_var[:] = np.arange(lon_dim)
                lat_var[:] = np.arange(lat_dim)
                precip_var[0, :, :] = precip3D_sum

                # Additional attributes can be added here

        else:
            print("No data available for concatenation.")

        # Move to the next day
        current_date += datetime.timedelta(days=1)

# Usage example
input_folder = r"D:\GPM\Final\New folder"
output_folder = r"D:\GPM\Final\Daily"
process_imerg_data("20220601", "20220607", input_folder, output_folder)
```
